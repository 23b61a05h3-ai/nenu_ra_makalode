```python
import numbers as np
import pandas as pd
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Convolution2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
x_train1=ImageDataGenerator(1./255)
x_test1=ImageDataGenerator(1./255)
```

```python
x_train=x_train1.flow_from_directory(r"C:\Users\HP\OneDrive\문서\TASK\Datasets\wines\train",batch_size=10,target_size=(64,64),class_mode='binary')
x_test=x_test1.flow_from_directory(r"C:\Users\HP\OneDrive\문서\TASK\Datasets\wines\test",batch_size=10,target_size=(64,64),class_mode='binary')
```

```python
model= Sequential()
```

```python
model.add(Convolution2D(32,(3,3),activation='relu',input_shape=(64,64,3)))
```

```python
model.add(Convolution2D(32,(3,3),activation='relu'))
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Convolution2D(32,(3,3),activation='relu'))
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
from tensorflow.keras.layers import Flatten
model.add(Flatten())
```

```python
model.add(Dense(units=128,activation='relu'))
```

```python
model.add(Dense(units=64,activation='relu'))
```

```python
model.add(Dense(units=32,activation='relu'))
```

```python
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,epochs=10,validation_data=x_test)
```
