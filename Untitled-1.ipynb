```python
import numpy as np 
import pandas as pd
import matplotlib as mat
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Viteesh\Downloads\task files\health.csv")
df.head()
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
 
le = LabelEncoder()
df.isnull().sum()
df.head()
```

```python
# Generate correlation matrix and visualize it
correlation_matrix = df.corr()
correlation_matrix
```

```python
x = df.drop(labels=["Outcome", "Insulin", "SkinThickness","BloodPressure"], axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
y = df["Outcome"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
lr = LogisticRegression()
knn = KNeighborsClassifier()
```

```python
dt.fit(x_train, y_train)
rf.fit(x_train, y_train)
lr.fit(x_train, y_train)
knn.fit(x_train, y_train)
le =dt.score(x_test, y_test), rf.score(x_test, y_test), lr.score(x_test, y_test), knn.score(x_test, y_test)
```

```python
lr_pred = lr.predict(x_test)
dt_pred = dt.predict(x_test)
rf_pred = rf.predict(x_test)
knn_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
lr_cm = confusion_matrix(y_test, lr_pred)
lr_cm
```

```python
dt_cm = confusion_matrix(y_test, dt_pred)
dt_cm
```

```python
rf_cm = confusion_matrix(y_test, rf_pred)
rf_cm
```

```python
knn_cm = confusion_matrix(y_test, knn_pred)
knn_cm
```

```python
from sklearn.metrics import accuracy_score

# Calculate accuracy for all models
lr_accuracy = accuracy_score(y_test, lr_pred)
dt_accuracy = accuracy_score(y_test, dt_pred)
rf_accuracy = accuracy_score(y_test, rf_pred)
knn_accuracy = accuracy_score(y_test, knn_pred)

# Display results
print("Model Accuracies:")
print(f"Logistic Regression: {lr_accuracy:.4f}")
print(f"Decision Tree: {dt_accuracy:.4f}")
print(f"Random Forest: {rf_accuracy:.4f}")
print(f"KNN: {knn_accuracy:.4f}")

# Create a summary dataframe
accuracy_df = pd.DataFrame({
    'Model': ['Logistic Regression', 'Decision Tree', 'Random Forest', 'KNN'],
    'Accuracy': [lr_accuracy, dt_accuracy, rf_accuracy, knn_accuracy]
})
print("\n", accuracy_df)
```
