```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
dd = km.fit_predict(dg)
```

```python
def clusters(dg,k_value):
    centroid_values = []
    k_values=[]
    for i in k_value:
        cs = KMeans(n_clusters=i)
        cs.fit_predict(dg)
        centroid_values.append(cs.inertia_)
        k_values.append(i)
    return centroid_values,k_values
        
```

```python
k=[1,2,3,4,5,6,7,8,9]
k
```

```python
j,s =clusters(dg,k)
```
